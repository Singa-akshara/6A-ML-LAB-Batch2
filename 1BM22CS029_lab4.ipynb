# Linear Regression

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt

# Load the data
data = pd.read_csv('canada_per_capita_income.csv')

# Prepare the data
X = data[['year']]  # Independent variable: Year
y = data['per capita income (US$)']  # Dependent variable: Per capita income

# Create and train the model
model = LinearRegression()
model.fit(X, y)

# Get the coefficient (m) and intercept (b)
m = model.coef_[0]  # Coefficient (slope)
b = model.intercept_  # Intercept (y-intercept)

# Print the values
print(f"Coefficient (m): {m:.2f}")
print(f"Intercept (b): {b:.2f}")

# Predict the per capita income in 2020
predicted_income = model.predict([[2020]])

# Print the prediction
print(f"Predicted per capita income in 2020: {predicted_income[0]:.2f} US$")

# (Optional) Visualize the data and regression line
plt.scatter(data['year'], data['per capita income (US$)'], color='red', marker='+')
plt.plot(data['year'], model.predict(data[['year']]), color='blue')  # Regression line
plt.xlabel('Year')
plt.ylabel('Per Capita Income (US$)')
plt.title('Canada Per Capita Income Prediction')
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt

# Load the housing area price data
df = pd.read_csv('housing_area_price.csv')

# Check for missing values in 'area' or 'price'
print(df.isnull().sum())  # Check if there are any missing values

# Handle missing values (choose one of the options below)

# Option 1: Drop rows with missing values
df = df.dropna(subset=['area', 'price'])

# Option 2: Impute missing values (replace with mean for example)
# df['area'].fillna(df['area'].mean(), inplace=True)
# df['price'].fillna(df['price'].mean(), inplace=True)

# Check the first few rows of the dataset to understand its structure
print(df.head())

# Plotting the data
plt.xlabel('Area (sqft)')
plt.ylabel('Price (US$)')
plt.scatter(df['area'], df['price'], color='red', marker='+')

# Create and train the linear regression model
X = df[['area']]  # Independent variable (area)
y = df['price']   # Dependent variable (price)

# Create and train the model
reg = linear_model.LinearRegression()
reg.fit(X, y)

# Plotting the regression line
plt.plot(df['area'], reg.predict(df[['area']]), color='blue')
plt.show()

# Predict the price of a house with a specific area
area_pred = reg.predict([[2000]])  # Example: Predict for a house with 2000 sqft area

print(f"Predicted Price for a house with 2000 sqft area: ${area_pred[0]:.2f}")

# Displaying the coefficient and intercept
print(f"Coefficient (m): {reg.coef_[0]:.2f}")
print(f"Intercept (b): {reg.intercept_:.2f}")

# Y = m * X + b (m is coefficient, b is intercept)
manual_price_prediction = 2000 * reg.coef_[0] + reg.intercept_
print(f"Manual Prediction for a house with 2000 sqft area: ${manual_price_prediction:.2f}")
```

# Multiple Regression

```python
import pandas as pd
import numpy as np
from sklearn import linear_model

# Load the hiring data
df = pd.read_csv('hiring.csv')

# Define a mapping for word to number conversions
word_to_number = {
    'one': 1,
    'two': 2,
    'three': 3,
    'four': 4,
    'five': 5,
    'six': 6,
    'seven': 7,
    'eight': 8,
    'nine': 9,
    'ten': 10,
    'eleven': 11,
    'twelve': 12,
    # Add more mappings as needed
}

# Convert 'experience' column to numeric, handling word values
df['experience'] = df['experience'].apply(lambda x: word_to_number.get(x, x))
df['experience'] = pd.to_numeric(df['experience'])

# Now you can fill NaN values with median safely
df['experience'] = df['experience'].fillna(df['experience'].median())

# Handle missing values in other columns as before
df['test_score(out of 10)'] = df['test_score(out of 10)'].fillna(df['test_score(out of 10)'].median())
df['interview_score(out of 10)'] = df['interview_score(out of 10)'].fillna(df['interview_score(out of 10)'].median())

# Create linear regression object
reg = linear_model.LinearRegression()

# Fit the model using relevant columns
reg.fit(df[['experience', 'test_score(out of 10)', 'interview_score(out of 10)']], df['salary($)'])

# Predict salaries for the given candidates
candidate1_salary = reg.predict([[2, 9, 6]])
candidate2_salary = reg.predict([[12, 10, 10]])

# Print the predicted salaries
print(f"Predicted salary for candidate 1 (2 yr experience, 9 test score, 6 interview score): ${candidate1_salary[0]:.2f}")
print(f"Predicted salary for candidate 2 (12 yr experience, 10 test score, 10 interview score): ${candidate2_salary[0]:.2f}")

# Display the coefficients and intercept (optional)
print(f"Coefficients: {reg.coef_}")
print(f"Intercept: {reg.intercept_}")
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
from sklearn.preprocessing import LabelEncoder

# Load the dataset
df = pd.read_csv('1000_Companies.csv')

# Display the first few rows of the dataset to understand its structure
print(df.head())

# Data Preprocessing:
# Encoding the 'State' column since it's categorical
label_encoder = LabelEncoder()
df['State'] = label_encoder.fit_transform(df['State'])

# Checking for any missing values and filling them (if necessary)
# Assuming 'Profit' column is the target and other columns are the features
df = df.fillna(df.median())  # Filling missing values with the median of the respective columns

# Independent variables (features) and dependent variable (target)
X = df[['R&D Spend', 'Administration', 'Marketing Spend', 'State']]  # Features
y = df['Profit']  # Target (Profit)

# Create the Linear Regression model
reg = linear_model.LinearRegression()

# Train the model
reg.fit(X, y)

# Coefficients and intercept of the trained model
print(f"Coefficients: {reg.coef_}")
print(f"Intercept: {reg.intercept_}")

# Predict the profit for the following candidate
# Given data: R&D Spend = 91694.48, Administration = 515841.3, Marketing Spend = 11931.24, State = 'Florida'

# Encoding the state 'Florida' into a numerical value
florida_state_encoded = label_encoder.transform(['Florida'])[0]

# Prediction for the given input
candidate_data = np.array([[91694.48, 515841.3, 11931.24, florida_state_encoded]])

# Predict the profit
predicted_profit = reg.predict(candidate_data)

print(f"Predicted Profit for the given candidate: ${predicted_profit[0]:.2f}")
```

```python
import pandas as pd
import numpy as np
from sklearn import linear_model

# Load the dataset
df = pd.read_csv('homeprices_Multiple_LR.csv')

# Display the first few rows of the dataset to understand its structure
print(df.head())

# Data Preprocessing: Check for and handle missing values
# Fill missing values with the median value of the respective columns
df = df.fillna(df.median())

# Independent variables (features) and dependent variable (target)
X = df[['area', 'bedrooms', 'age']]  # Features: area, bedrooms, age
y = df['price']  # Target: price

# Create the Linear Regression model
reg = linear_model.LinearRegression()

# Train the model
reg.fit(X, y)

# Coefficients and intercept of the trained model
print(f"Coefficients: {reg.coef_}")
print(f"Intercept: {reg.intercept_}")

# Predict the price for the following new house:
# Given data: area = 3000, bedrooms = 3, age = 40
new_house = np.array([[3000, 3, 40]])

# Predict the price
predicted_price = reg.predict(new_house)

print(f"Predicted Price for the given house (area=3000, bedrooms=3, age=40): ${predicted_price[0]:.2f}")
```
